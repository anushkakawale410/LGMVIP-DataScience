### Image to Pencil Sketch with Python


```python
#Importing libraries
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Loading the image
img = cv2.imread("E:\internship task\dog-img.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
#Converting Image to Gray Scale
image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(image_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
#Inverting the Image
image_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.imshow(image_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
#Smoothing the Image
smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
#Pencil sketch
sketch = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(sketch,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
